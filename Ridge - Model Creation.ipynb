```python
import pandas as pd
```

```python
data=pd.read_csv('50_Startups.csv')
data
```

```python
data=pd.get_dummies(data,dtype=int,drop_first=True)
```

```python
data
```

```python
data.columns
```

```python
inde=data[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
inde
```

```python
dep=data[['Profit']]
dep
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
param_grid={ 'solver':['auto','svd','cholesky','lsqr','sparse_cg','sag','saga'],
            'alpha'  :[1,10,100]}
```

```python
grid=GridSearchCV(Ridge(),param_grid,refit=True,verbose=3,n_jobs=-1)
```

```python
grid.fit(inde,dep)
```

```python
re=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
print(grid.best_params_)
```

```python
import pickle
Filename='Final Ridge - Gridsearch CV.sav'
pickle.dump(grid,open(Filename,'wb'))
```

```python
load=pickle.load(open('Final Ridge - Gridsearch CV.sav','rb'))
result=load.predict([[24,17,0,0,1]])
```
